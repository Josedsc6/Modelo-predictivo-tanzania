Renombro los valores de las variables para que todas sean minúsculas, pero comiencen con mayúsculas, en ambos dataframes, ya que he observado que algunas variables tienen el mismo nombre pero una en minúscula y otras en mayúsculas.

```python
train_values = train_values.applymap(lambda x: x.capitalize() if isinstance(x, str) else x)
test_values =test_values.applymap(lambda x: x.capitalize() if isinstance(x, str) else x)
```

El siguiente paso es unir los datos de entrenamiento con las etiquedas mediante un .merge teniendo como punto de unión los 'id'. También guardo los 'id' que me servirán más tarde para generar el archivo de submission a evaluar.

```python
train_data = pd.merge(train_values, train_labels, on='id')
test_data = test_values.copy()
test_ids = test_data['id']
```

Veo que la variable Recorded_by sólo tiene 1 valor en train_values por lo que no me da nada de información y la borro. También observo que en public_meeting los False vienen enmascarados como NaN, por lo que hago este cambio

```python
train_data['public_meeting'].value_counts()
```

```python
true_count = train_data['public_meeting'].value_counts().get(True, 0)
nan_count = train_data['public_meeting'].isna().sum()

print(f"True: {true_count}")
print(f"NaN: {nan_count}")
```

```python
train_data = train_data.drop(columns=['recorded_by'])
test_data = test_data.drop(columns=['recorded_by'])

train_data['public_meeting'] = train_data['public_meeting'].fillna(False)
test_data['public_meeting'] = test_data['public_meeting'].fillna(False)
```

Después de verificar similitudes entre distintas parejas o grupos de variables, veo que region_code tiene 27 valores distintos, mientras que region tiene 21. Las comparo y veo que algunas regiones tienen varios códigos por lo que la información de region_code está contenida en region, así que la borro

```python
train_data['region'].nunique()
```

```python
train_data['region_code'].nunique()
```

```python
# Identificar las regiones que tienen más de un código
region_code_duplicates = train_data.groupby('region')['region_code'].nunique()
regions_with_multiple_codes = region_code_duplicates[region_code_duplicates > 1]

regions_with_multiple_codes

train_data = train_data.drop(columns=['region_code'])
test_data = test_data.drop(columns=['region_code'])
```

Después de muchas pruebas, veo que lo mejor para el preprocesado de datos es quedarnos con las demás variables (después haremos otra limpieza más general), imputar los valores faltantes por la moda (para categóricas) y la mediana (para numéricas)

```python
categorical_columns = train_data.select_dtypes(include='object').columns.drop('status_group')
for column in categorical_columns:
    train_data[column] = train_data[column].fillna(train_data[column].mode()[0])
    test_data[column] = test_data[column].fillna(test_data[column].mode()[0])

numerical_columns = train_data.select_dtypes(include=['float64', 'int64']).columns
for column in numerical_columns:
    train_data[column] = train_data[column].fillna(train_data[column].median())
    test_data[column] = test_data[column].fillna(test_data[column].median())
```

Elimino la variable 'id' que no me aporta información y ya los tengo guardados en test_ids. En cuanto a la variable 'scheme_name', la borro también ya que tiene 28810 missings (la mitad de los datos aproximadamente)

```python
train_data.drop(columns=['scheme_name', 'id'], inplace=True, errors='ignore')
test_data.drop(columns=['scheme_name', 'id'], inplace=True, errors='ignore')
```

Paso las fechas a datetime mediante una función y borro la variable original. Con los años de esta fecha creo una variable nueva que incluya la antigüedad (recorded - construction)

```python
def procesar_fechas(data):
    data['date_recorded'] = pd.to_datetime(data['date_recorded'])
    data['recorded_year'] = data['date_recorded'].dt.year
    data['recorded_month'] = data['date_recorded'].dt.month
    data['recorded_day'] = data['date_recorded'].dt.day
    data.drop(columns=['date_recorded'], inplace=True)
    return data

train_data = procesar_fechas(train_data)
test_data = procesar_fechas(test_data)
```

```python
median_year = train_data.loc[train_data['construction_year'] > 0, 'construction_year'].median()
train_data['construction_year'] = train_data['construction_year'].replace(0, median_year)
test_data['construction_year'] = test_data['construction_year'].replace(0, median_year)
train_data['waterpoint_age'] = train_data['recorded_year'] - train_data['construction_year']
test_data['waterpoint_age'] = test_data['recorded_year'] - test_data['construction_year']
```

En cuando a la geografía de cada bomba, las agrupo en 10 clusters mediante KMeans basándome en la latitud y longitud. Creo la variable geo_cluster que incluya esta información

```python
coords = pd.concat([train_data[['latitude', 'longitude']], test_data[['latitude', 'longitude']]])
kmeans = KMeans(n_clusters=10, random_state=42).fit(coords)
train_data['geo_cluster'] = kmeans.predict(train_data[['latitude', 'longitude']])
test_data['geo_cluster'] = kmeans.predict(test_data[['latitude', 'longitude']])
```

El siguiente paso es separar la variable objetivo de los datos de entrenamiento y eliminar las columnas categóricas que tengan demasiadas categorías (alta cardinalidad), para poder convertirlas a dummies y que no generen demasiadas columnas.

```python
y = train_data['status_group']
train_data.drop(columns=['status_group'], inplace=True)

cat_cols = train_data.select_dtypes(include='object').columns
high_card_cols = [col for col in cat_cols if train_data[col].nunique() > 100]
train_data.drop(columns=high_card_cols, inplace=True)
test_data.drop(columns=high_card_cols, inplace=True, errors='ignore')

train_data = pd.get_dummies(train_data, drop_first=True)
test_data = pd.get_dummies(test_data, drop_first=True)
```

Alineo las columnas del test para que tenga las mismas columnas del train, rellenando con ceros si faltan algunas. Se escalan las variables numéricas para mejorar el rendimiento del modelo

```python
test_data = test_data.reindex(columns=train_data.columns, fill_value=0)
```

```python
X = train_data
scaler = StandardScaler()
X = scaler.fit_transform(X)
test_data = scaler.transform(test_data)
```

# Una vez finalizado el procesamiento de los datos, pasamos al modelado predictivo

Optimizamos el modelo mediante RandomizedSearchCV en el que incluimos varios parámetros a explorar y después de probar muchas combinaciones cruzadas nos elige la mejor, para a continuación predecir los resultados mediante RandomForest (el modelo que me ha dado mayor accuracy).

###### Tarda un poquito en ejecutar, uso el n_jobs=-1 para que aproveche toda la CPU de mi ordenador (en mi caso tarda unos 40 segundos)

```python
model = RandomForestClassifier(random_state=42, n_jobs=-1)
param_dist = {
    'n_estimators': [100, 200, 300],
    'max_depth': [20, 30, None],
    'min_samples_split': [2, 5, 10],
    'min_samples_leaf': [1, 2, 4]
}
random_search = RandomizedSearchCV(model, param_dist, n_iter=10, cv=3, random_state=42, n_jobs=-1)
random_search.fit(X, y)
```

```python
best_model = random_search.best_estimator_

y_test_pred = best_model.predict(test_data)
```

### Por último generamos el archivo a evaluar que tenga el formato especificado en el enunciado y lo probamos en la pagina web del concurso

```python
submission = pd.DataFrame({
    'id': test_ids,
    'status_group': y_test_pred
})
submission.to_csv('Submission_Jose_Manuel_Jimenez_Serrano.csv', index=False)
```
